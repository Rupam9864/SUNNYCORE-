```python
import os
import regex
import datetime
import pandas as pd
import numpy as np
from moviepy import AudioFileClip #moviepy is not working properly, need other method
```

```python
textfilepath = 'C:/Users/rajt8/Desktop/SunnyCore_Data/datasets/raw/MELD/train/'
textfilename = 'train_sent_emo.csv'
vediofilepath = 'C:/Users/rajt8/Desktop/SunnyCore_Data/datasets/raw/MELD/train/train_splits/'
audiofilepath = 'C:/Users/rajt8/Desktop/SunnyCore_Data/datasets/raw/MELD/train/audio/'
audiofileformat = '.wav'
```

```python
df=pd.read_csv(textfilepath+textfilename)
df
```

```python
vediolist=os.listdir(vediofilepath)
vediolist=pd.DataFrame(vediolist)
vediolist.columns=['Filename']
vediolist=pd.concat([vediolist,pd.DataFrame(columns=['Majorrank','Minorrank'])])
```

```python
vediolist
```

```python
vediolist.iloc[2,0]
```

```python
for i in range(len(vediolist)):
    text=vediolist.iloc[i,0]
    Majorranktext=text.split('_')[0]
    Minorranktext=text.split('_')[1]

    Majorrankint=regex.findall(r'\d+',Majorranktext)[0]
    Minorrankint=regex.findall(r'\d+',Minorranktext)[0]

    vediolist.loc[i,'Majorrank']=int(Majorrankint)
    vediolist.loc[i,'Minorrank']=int(Minorrankint)
```

```python
vediolist
```

```python
vediolist.sort_values(by=['Majorrank','Minorrank'],inplace=True)
```

```python
vediolist
```

```python
vediolistproceeded=vediolist['Filename'].to_list()
vediolistproceeded[9988]
```

```python
vediolistproceeded
```

```python
import os
import subprocess

# Ensure the output directory exists
if not os.path.exists(audiofilepath):
    os.makedirs(audiofilepath)

for i in range(len(vediolistproceeded)):
    vediofilename = vediolistproceeded[i]
    filename = vediofilename.split('.')[0]
    audiofilename = filename + audiofileformat
    input_path = os.path.join(vediofilepath, vediofilename)
    output_path = os.path.join(audiofilepath, audiofilename)
    
    print(f"Processing file {i + 1}/{len(vediolistproceeded)}: {vediofilename}")
    try:
        # Run FFmpeg command to extract audio
        command = [
            "ffmpeg",
            "-i", input_path,
            "-vn",  # Disable video
            "-acodec", "pcm_s16le",  # Use WAV format
            output_path
        ]
        subprocess.run(command, check=True)
        print(f"Successfully processed: {audiofilename}")
    except subprocess.CalledProcessError as e:
        print(f"Error processing file {vediofilename}: {e}")
        with open("error_log.txt", "a") as log:
            log.write(f"Error with {vediofilename}: {e}\n")
```
