```python
#import os
import re
#import regex
#import datetime
#import subprocess
#import sys
import pandas as pd
import numpy as np
#import ffmpeg
#from matplotlib import pyplot as plt
```

```python
textfilepath = 'C:/data/raw/MELD/train/'
textfilename = 'train_sent_emo.csv'
```

```python
df=pd.read_csv(textfilepath+textfilename)
```

```python
def timecal(inputstring):
    hour,min,sec,msec=re.split(':|,',inputstring)
    hourins=int(hour)*3600
    minins=int(min)*60
    second=int(sec)
    milisec=int(msec)*10e-4
    timeInSec=hourins+minins+second+milisec
    return timeInSec

def wordcounter(inputString):
    inputlist=inputString.split(' ')
    sentencelength=len(inputlist)
    return sentencelength
```

```python
df[['StarttimeInSec','EndTimeinsec','deltaT','num of word','wordPerSec']]=None
```

```python
for i in range(len(df)):
    starttime=df.loc[i,'StartTime']
    endtime=df.loc[i,'EndTime']
    textValue=df.loc[i,'Utterance']

    starttime=timecal(starttime)
    endtime=timecal(endtime)


    df.loc[i,'StarttimeInSec']=starttime
    df.loc[i,'EndTimeinsec']=endtime
    df.loc[i,'deltaT']=endtime-starttime
    df.loc[i,'num of word']=wordcounter(textValue)
    df.loc[i,'wordPerSec']=wordcounter(textValue)/(endtime-starttime)
```

```python
dfcleaned=df.loc[df['wordPerSec']<4.5] 
#4.5 (Upward approximation 265/60) is based on the average word per second which a adult can deal with when they are glancing.
```
