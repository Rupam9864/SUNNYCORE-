```python
#import os
#import re
#import regex
#import datetime
#import subprocess
#import sys
import pandas as pd
#import numpy as np
#import ffmpeg
from matplotlib import pyplot as plt
```

```python
#please put the path of dataset here
filename="C:/data/proceeded/text/textdata.csv"
```

```python
df=pd.read_csv(filename)
```

```python
df
```

```python
numOfWord=df['num of word'].value_counts().sort_index()

plt.figure(dpi=500,figsize=(25,15))
plt.bar(numOfWord.index,numOfWord.values)

plt.xlabel('Quantity of word')
plt.ylabel('Quantity of data')
```

```python
Sentiment=df['Sentiment'].value_counts()
plt.figure(dpi=500,figsize=(10,10))
plt.pie(Sentiment,autopct='%1.1f%%')
plt.legend(Sentiment.index)
```

```python
Emotion=df['Emotion'].value_counts()
plt.figure(dpi=500,figsize=(25,15))
plt.bar(Emotion.index,Emotion.values)

plt.xlabel('types of emotion')
plt.ylabel('Quantity of emotion')
```
