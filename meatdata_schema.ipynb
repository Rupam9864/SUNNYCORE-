```python
import pandas as pd
import os

# Paths to datasets
PROCESSED_DATA_PATH = 'datasets\processed\meld_features_updated.csv'
METADATA_PATH = 'datasets\metadata\dataset_inventory.csv'

# Licensing Information
licensing_info = {
    "MELD": "https://github.com/SenticNet/MELD",
    "IEMOCAP": "https://sail.usc.edu/iemocap/",
    "CREMA-D": "https://github.com/CheyneyComputerScience/CREMA-D",
    "RAVDESS": "https://zenodo.org/record/1188976",
    "GoEmotions": "https://github.com/google-research/google-research/tree/master/goemotions"
}

# Preprocessing Steps Applied
preprocessing_steps = [
    "Text: Tokenization, lowercasing, special character removal, stopword removal, lemmatization.",
    "Audio: Resampling to 16kHz, noise reduction using spectral gating, MFCC feature extraction.",
    "Multimodal: Temporal alignment of text and audio, prosodic feature extraction (pitch, energy)."
]

# Load Processed Data
data = pd.read_csv(PROCESSED_DATA_PATH)

# Calculate Dataset Statistics
dataset_stats = {
    "Total Samples": len(data),
    "Class Distribution": data['Emotion'].value_counts().to_dict(),
    "Missing Values": data.isnull().sum().to_dict(),
    "Duplicates": data.duplicated().sum()
}

# Create Metadata DataFrame
metadata = {
    "Aspect": ["Dataset Statistics", "Licensing Information", "Data Quality Metrics", "Preprocessing Steps"],
    "Details": [
        str(dataset_stats),
        str(licensing_info),
        f"Missing Values: {dataset_stats['Missing Values']}, Duplicates: {dataset_stats['Duplicates']}",
        "\n".join(preprocessing_steps)
    ]
}

metadata_df = pd.DataFrame(metadata)

# Save Metadata to CSV
os.makedirs(os.path.dirname(METADATA_PATH), exist_ok=True)
metadata_df.to_csv(METADATA_PATH, index=False)
print(f"Metadata schema saved to {METADATA_PATH}!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot Class Distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=data['Emotion'].value_counts().index, y=data['Emotion'].value_counts().values, palette='viridis')
plt.title('Class Distribution', fontsize=16)
plt.xlabel('Emotion', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('datasets/metadata/class_distribution.png')  # Save chart
plt.show()
```

```python
# Plot Missing Values
missing_values = data.isnull().sum()
plt.figure(figsize=(8, 5))
sns.barplot(x=missing_values.index, y=missing_values.values, palette='coolwarm')
plt.title('Missing Values per Column', fontsize=16)
plt.xlabel('Columns', fontsize=14)
plt.ylabel('Missing Count', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('datasets/metadata/missing_values.png')  # Save chart
plt.show()
```
